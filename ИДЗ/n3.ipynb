```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
class Options:
    class call:
        def buy(self,X,E,price):
            L=[]
            for x in X:
                if(x<=E):
                    L.append(-price)
                else:
                    L.append(x-E-price)
            return np.array(L)

        def sell(self,X,E,price):
            L=[]
            for x in X:
                if(x<=E):
                    L.append(price)
                else:
                    L.append(price+E-x)
            return np.array(L)
            
    class put:
        def buy(self,X,E,price):
            L=[]
            for x in X:
                if(x<=E):
                    L.append(E-x-price)
                else:
                    L.append(-price)
            return np.array(L)

        def sell(self,X,E,price):
            L=[]
            for x in X:
                if(x<=E):
                    L.append(x-E-price)
                else:
                    L.append(-price)
            return np.array(L)
```

```python
x=45
y=5
z=6
```

```python
X=np.arange(0,2*x,0.001)
opt1 = Options.call()
plt.figure(figsize=(10, 4.5))
plt.grid()
plt.plot(X,opt1.buy(X,x,0),label='call')
plt.legend()
plt.show()
```

```python
X=np.arange(0,2*x,0.001)
opt = Options.put()
plt.figure(figsize=(10, 4.5))
plt.grid()
plt.plot(X,opt.buy(X,x,0),label='put')
plt.legend()
plt.show()
```
